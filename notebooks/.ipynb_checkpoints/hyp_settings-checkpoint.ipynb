```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Doing For Only 1 STATION
data = pd.read_json("../results/hyperp_data_frequency.json")
data
```

```python
poly_degree = []
alpha = []
interaction_only = []
include_bias = []

for i in range(0, 240, 8):
    poly_degree.append(data["best params"][i]['poly_features__degree'])
    alpha.append(data["best params"][i]['ridge_reg__alpha'])
    interaction_only.append(data["best params"][i]['poly_features__interaction_only'])
    include_bias.append(data["best params"][i]['poly_features__include_bias'])

print(poly_degree)
print(alpha)
print(interaction_only)
print(include_bias)
```

```python
df = pd.DataFrame({'Polynomial Degree': poly_degree})
df.groupby('Polynomial Degree').size().plot(kind='bar')
plt.show()
```

```python
df = pd.DataFrame({'Alpha': alpha})
df.groupby('Alpha').size().plot(kind='bar')
plt.show()
```

```python
df = pd.DataFrame({'Interaction': interaction_only})
df.groupby('Interaction').size().plot(kind='bar')
plt.show()
```

```python
df = pd.DataFrame({'Include Bias': include_bias})
df.groupby('Include Bias').size().plot(kind='bar')
plt.show()
```
